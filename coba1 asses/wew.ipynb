```python
# create map and display it
lon_map = folium.Map(location=[latitude, longitude], zoom_start=15)

# display the map of London
lon_map
```

```python
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df_lon.Latitude, df_lon.Longitude):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
# add pop-up text to each marker on the map
latitudes = list(df_lon.Latitude)
longitudes = list(df_lon.Longitude)
labels = list(df_lon.CrimeType)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(lon_map)   
    
# add incidents to map
lon_map.add_child(incidents)
```

```python
# create map and display it
lon_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# loop through the 600 crimes and add each to the map
for lat, lng, label in zip(df_lon.Latitude, df_lon.Longitude, df_lon.CrimeType):
    folium.CircleMarker(
        [lat, lng],
        radius=5, # define how big you want the circle markers to be
        color='yellow',
        fill=True,
        popup=label,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(lon_map)

# show map
lon_map
```

```python
from folium import plugins

# let's start again with a clean copy of the map of London
lon_map = folium.Map(location = [latitude, longitude], zoom_start = 12)

# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(lon_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(df_lon.Latitude, df_lon.Longitude, df_lon.CrimeType):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

# display map
lon_map
```
